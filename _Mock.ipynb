```python
import pandas as pd

data = {
    'id': [1, 2, 3, 4], 
    'age': [25, 30, 35, 40],
    'gender': ['Male', 'Female', 'Male', 'Female'],
    'city': ['Delhi', 'Mumbai', 'Kolkata', 'Chennai']
}

df = pd.DataFrame(data)

# 1st dataframe

df1 = df[['id', 'age', 'gender']].iloc[:2]
df1
```

```python
# 2nd dataframe

df2 = df[['id', 'age', 'city']].iloc[:2] 
df2
```

```python
# merge dataframe

merged_df =  pd.merge(df1,df2, on = 'id')
merged_df
```

```python
# import to csv file 

merged_df.to_csv("merged_dataframe.csv", index=False)
merged_df
```
