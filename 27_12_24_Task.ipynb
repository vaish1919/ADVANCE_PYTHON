## Find the average salary for each department in a company.

```python
import pandas as pd
```

```python
data = {'Department':['HR','IT','HR','IT','Finance'],'Salary':[40000,50000,45000,55000,60000]}
df = pd.DataFrame(data)

result = df.groupby('Department')['Salary'].mean()
result
```

##  Count the number of employees in each department.

```python
result = df.groupby('Department')['Salary'].count()
result
```

## Calculate the total salary paid in each department.

```python
result = df.groupby('Department')['Salary'].sum()
result
```

```python
result = df.groupby('Department')['Salary'].agg(['mean','max'])
result
```

##  Find the total salary for each combination of department and job role.

```python
data = {'Department': ['HR', 'IT', 'HR', 'IT', 'Finance'],
        'Role': ['Manager', 'Developer', 'Assistant', 'Developer', 'Analyst'],
        'Salary': [40000, 50000, 45000, 55000, 60000]}

df = pd.DataFrame(data)

result = df.groupby(['Department', 'Role'])['Salary'].sum()
result
```

## Retrieve the first employee record from each department.

```python
result = df.groupby('Department').first()
result
```

## Find departments where the total salary is greater than 90,000.

```python
result = df.groupby('Department')['Salary'].sum()
filtered_result = result[result > 90000]
filtered_result
```

## Add a column that shows the cumulative salary in each department.

```python
df['Cumulative_Salary'] = df.groupby('Department')['Salary'].cumsum()
df
```

## Normalize salaries within each department by subtracting the group mean.

```python
df['Normalized_Salary'] = df.groupby('Department')['Salary'].transform(lambda x: x - x.mean())
df
```

## Find out how many employees are there in each department.

```python
result = df.groupby('Department').size()
result
```
