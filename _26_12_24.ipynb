### Groupby questions 

###  How to group by multiple columns?



```python
data = {'Category': ['A', 'A', 'B', 'B', 'A'],
        'Subcategory': ['X', 'Y', 'X', 'Y', 'X'],
        'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

grouped = df.groupby(['Category', 'Subcategory']).sum()
grouped
```

###  How do you use groupby in pandas?

```python
data = {
        'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
         'Values'  : [10, 20, 15, 25, 10, 30]}
df = pd.DataFrame(data)

grouped = df.groupby('Category')
grouped.sum()
```

###  How to apply multiple aggregation functions?

```python
grouped = df.groupby('Category').agg(['mean', 'sum'])
grouped
```
